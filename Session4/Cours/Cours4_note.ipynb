Attention, il faut mettre le tilde pour avoir les emails qui ne sont pas au même format que celui du regex.

le regex ici représente : '.+@[A-Za-z0-9_\-.]+\.[A-Za-z]{2,}'

ici, le regex nous donne les emails valides avec @"caractères"."2 lettres minimum"

le tilde nous donne le contraire de l'ensemble identifié par le regex

Pour les séries de pandas, on utilise : 
    - tilde pour la négation 
    - & pour et
    - | pour ou 

```python
#email avec des aliases 

df.loc[df['email'].str.contains('\+'), 'email'].values
```

```python
#pour supprimer les alias 

tab_email = df['email'].str.extract('([^+]+)(\+.*)?(@.+)').sort_values(0)
tab_email
```

ici on a : '([^+]+)(\+.*)?(@.+)'
    
avec: 
- '([^+]+)représentant extraction 0
- (\+.*)représentant extraction 1
- ?(@.+)' représentant extraction 2

```python
df['email']=tab_email[0] + tab_email[2]
```

```python
df.head()
```

```python
df['email'] = tab_email[0] + tab_email[2]
df.sort_values('email')
```
